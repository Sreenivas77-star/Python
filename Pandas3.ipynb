```python
import numpy as np
import pandas as pd
```

```python
num_samples = 20

age = np.random.randint(18, 65, num_samples)
salaries = np.random.randint(20000, 100000, num_samples)
gender = np.random.choice(['Male', 'Female'], num_samples)
employee_id = np.random.randint(10000000,200000000, num_samples)
dept = np.random.choice(['HR', 'IT', 'Finance', 'Marketing'], num_samples)
```

```python
data_dict = {
  "Age" : age,
  "Salary" : salaries,
  "Gender": gender,
  "EmployeeID": employee_id,
  "Dept": dept
}
```

```python
data = pd.DataFrame(data_dict)
data
```

```python
data.isnull()
```

```python
data.dropna(inplace = True)
data
```
