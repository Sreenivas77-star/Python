```python
import matplotlib.pyplot as plt

plt.hist(df1['Marks'], bins=10, color='orange') # You can adjust the number of bins and choose a different color
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.title('Distribution of Marks') # Optional: Add a title
plt.show()
```

```python
marks_counts = df1['Marks'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
marks_counts.plot(kind='bar', color='Red')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.title('Number of Students per Mark')
plt.xticks(rotation=0) # Keep labels horizontal for readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```
